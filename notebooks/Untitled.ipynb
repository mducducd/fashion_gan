```python
import os
import glob
import numpy as np
import random
import matplotlib.pyplot as plt
from PIL import Image

from utils import search_utils

%matplotlib inline
```

```python
feature_gen = search_utils.ResnetFeatureGenerator()
```

```python
product_path = '../../data/fashion/dresses/'
features_path = './data/features/fashion/dresses/resnet/'
search = search_utils.Search(product_path, features_path, feature_gen)
```

```python
f = search.features
f = f + np.abs(f.min())
f = f / f.max() * 255
```

```python
m = f.mean(axis=0).reshape(1, 2048)
m = m.repeat(100, axis=0)
```

```python
fig, ax = plt.subplots(figsize=(100,100))
ax.matshow(m, cmap='gray')
```

```python
fig, ax = plt.subplots(figsize=(100,100))
s = f[:200].copy()
s[s < 181] = 0
ax.matshow(s, cmap='gray')
```
